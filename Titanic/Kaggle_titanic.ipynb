```python
#Importando as bibliotecas

import pandas as pd
import numpy as np
```

```python
#Carregando o dataset

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
# Verificando dados nulos
train.isnull().sum()
```

```python
# Tratando os dados nulos substituindo por -1
train.fillna(-1, inplace=True)
train.isnull().sum()
```

```python
# Colunas do dataset
train.columns
```

```python
# Transformando a coluna 'Sex' em uma numerica
def transformar_sexo(valor):
    if valor == 'female':
        return 1
    else:
        return 0
    
train['Sex_binary'] = train['Sex'].map(transformar_sexo)
```

```python
#Verificando as correlacoes das variaveis
train.corr().round(4)
```

```python
# Separando as variaveis que vamos usar
variaveis = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_binary']
```

```python
# Serando os dados
X = train[variaveis]
y = train['Survived']
```

```python
# Separando 80% para treino e 20 para validacao
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Usaremos o modelo RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

#Instanciando o modelo
modelo = RandomForestClassifier(n_estimators=100, n_jobs=1, random_state = 0)

#Treinando o modelo
modelo.fit(X_train, y_train)
```

```python
# Usando os dados de testes no modelo
p = modelo.predict(X_test)
```

```python
# Calculando o resultado do modelo
np.mean(y_test == p)
```

```python
# Medindo o modelo com diversos dados

from sklearn.model_selection import RepeatedKFold

resultados = []
kf = RepeatedKFold(n_splits=2, n_repeats=5, random_state = 10)
for linhas_treinos, linhas_valid in kf.split(X):
    print('Linhas Treinos', linhas_treinos.shape[0])
    print('Linhas Validacao', linhas_valid.shape[0])

    X_train, X_valid = X.iloc[linhas_treinos], X.iloc[linhas_valid]
    y_train, y_valid = y.iloc[linhas_treinos], y.iloc[linhas_valid]

    modelo = RandomForestClassifier(n_estimators=100, n_jobs=1, random_state=0)
    modelo.fit(X_train, y_train)
    p = modelo.predict(X_valid)

    acc = np.mean(p == y_valid)
    resultados.append(acc)
    print('Acc:', acc)
    print('')
    #print(X_treino.head())

print('Acurácia:', np.mean(resultados))
```

```python
# Testando a LogisticRegression
from sklearn.linear_model import LogisticRegression

#Instanciando o modelo
modelo = LogisticRegression()

#Treinando o modelo
modelo.fit(X_train, y_train)
```

```python
# Usando os dados de testes no modelo
p = modelo.predict(X_test)
```

```python
# Calculando o resultado do modelo
np.mean(y_test == p)
```

```python
#Comparando os resultados
X_valid_check = train.iloc[y_test.index].copy()
X_valid_check['p'] = p

erros = X_valid_check[X_valid_check['Survived'] != p]
print('Quantidade de erros:', erros.shape[0])
erros = erros[['PassengerId','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked','Sex_binary','p','Survived']]
erros
```

```python
# Adicionando novas variaveis
train['Embarked_S'] = (train['Embarked'] == 'S').astype(int)
train['Embarked_C'] = (train['Embarked'] == 'C').astype(int)

train['Cabine_nula'] = (train['Cabin'].isnull()).astype(int)

train['Name_contains_miss'] = (train['Name'].str.contains('Miss')).astype(int)
train['Name_contains_mrs'] = (train['Name'].str.contains('Mrs')).astype(int)
train['Name_contains_master'] = (train['Name'].str.contains('Master')).astype(int)
train['Name_contains_col'] = (train['Name'].str.contains('Col')).astype(int)
train['Name_contains_major'] = (train['Name'].str.contains('Major')).astype(int)
train['Name_contains_mr'] = (train['Name'].str.contains('Mr')).astype(int)
```

```python
# Verificando novo dataset
print(train.shape)
train.head()
```

```python
#Separando os dados
novas_variaveis = ['Pclass', 'Sex_binary', 'Age', 'SibSp', 'Parch', 'Fare',
            'Name_contains_miss', 'Name_contains_mrs', 'Name_contains_master',
            'Name_contains_col', 'Name_contains_major', 'Name_contains_mr']

X = train[novas_variaveis]
y = train['Survived']
```

```python
#Separando os dados de testes e treinos
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
# Testando com RandomForestClassifier

#Instanciando o modelo
modelo = RandomForestClassifier(n_estimators=100, n_jobs=1, random_state=0)

#Treinando o modelo
modelo.fit(X_train, y_train)

#Usando o modelo
p = modelo.predict(X_test)

np.mean(p == y_test)
```

```python
# Testando com LogisticRegression

#Instanciando o modelo
modelo = LogisticRegression()

#Treinando o modelo
modelo.fit(X_train, y_train)

#Usando o modelo
p = modelo.predict(X_test)

np.mean(p == y_test)
```

```python
#Gerando arquivo com os dados de test

#Tratando os dados
test.fillna(-1, inplace=True)
test['Sex_binary'] = test['Sex'].map(transformar_sexo)

test['Embarked_S'] = (test['Embarked'] == 'S').astype(int)
test['Embarked_C'] = (test['Embarked'] == 'C').astype(int)

test['Cabine_nula'] = (test['Cabin'].isnull()).astype(int)

test['Name_contains_miss'] = (test['Name'].str.contains('Miss')).astype(int)
test['Name_contains_mrs'] = (test['Name'].str.contains('Mrs')).astype(int)
test['Name_contains_master'] = (test['Name'].str.contains('Master')).astype(int)
test['Name_contains_col'] = (test['Name'].str.contains('Col')).astype(int)
test['Name_contains_major'] = (test['Name'].str.contains('Major')).astype(int)
test['Name_contains_mr'] = (test['Name'].str.contains('Mr')).astype(int)

X = test[novas_variaveis]

p = modelo.predict(X)

sub = pd.Series(p, index = test['PassengerId'], name = 'Survived')
sub.to_csv('result.csv', header = True)
```
